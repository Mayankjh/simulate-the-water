```python
import pandas as pd
data = open('rest.csv','r',encoding="utf8")
print(data)
p = pd.read_csv(data)
p.head()
#p['Name']
# print(data.read(69532)[69512-40:69512+10])
```

```python
p.Name=p.Name.fillna('Afghanistan')
```

```python
n=p['Name']
l=[]
for name in n:
    l.append(name)
m=list(set(l))
```

```python
from geopy.geocoders import Nominatim
lat=[]
lon=[]
for names in l:
    geolocator = Nominatim(user_agent="specify_your_app_name_here")
    location = geolocator.geocode(names,timeout=1000000)
    #print((location.latitude, location.longitude))
    if location:
        lat.append(location.latitude)
        lon.append(location.longitude)
    else:
        lat.append(33.7680065)
        lon.append(66.2385139)
        
```
